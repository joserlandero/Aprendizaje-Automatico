## Prueba de hipótesis de correlación

```python
from scipy.stats import pearsonr

# Correlación entre café y sueño
r, p = pearsonr(df['Coffee_Intake'],df['Physical_Activity_Hours'])

print(f"Coeficiente de correlación: {r:.3f}")
print(f"Valor p: {p:.4f}")

if p < 0.05:
    print("→ Rechazamos H₀: la correlación es estadísticamente significativa.")
else:
    print("→ No se rechaza H₀: la correlación no es significativa.")

    r, p = pearsonr(df['Caffeine_mg'],df['Physical_Activity_Hours'])

print(f"Coeficiente de correlación: {r:.3f}")
print(f"Valor p: {p:.4f}")

if p < 0.05:
    print("→ Rechazamos H₀: la correlación es estadísticamente significativa.")
else:
    print("→ No se rechaza H₀: la correlación no es significativa.")
```
