```python
import pandas as pd
```

Se usará una base de datos relacionada al cafe y el impacto en la salud del usuario

```python
data = pd.read_csv('synthetic_coffee_health_10000.csv')
```

```python
#Imprimimos los primeros 10 registros del dataframe
data.head(10)
```

```python
#Informacion del dataframe
data.info()
```

```python
#Descripción mas detallada de las columnas
data.describe()
```

```python
#Vamos a excluir la columna BMI ya q siento que no aporta
data = data.drop('BMI',axis=1)
data.head(1)
```

```python
#Vamos a filtrar registros de mexico a ver si hay y cuantos registros som
data_filtrado = data[data['Country']=='Mexico']
data_filtrado.shape[0]
#Observamos que son 483 registros
```

```python
#Ahora regisrros de mexico con sleep houers menor a 7
data_filtrado = data[(data['Country']=='Mexico') & (data['Sleep_Hours'] < 7)]
data_filtrado.shape[0]
#Observamos que son 285 registros
```

```python
#Ahora agrupamos y contamos cuantos tienen x Streess_Level
data_filtrado['Stress_Level'].value_counts()
#Observamos que tenemos mas registros con poco nivel de estress
```

```python
#Del original filtramos registros nulos
data.isnull().sum()
#Observamos que health Issues es la que ams tiene aunq eso lo vimos al inicio con el info()
```

```python
#Obtenemos los dummies del stress level
df_stress = pd.get_dummies(data['Stress_Level'])
```

```python
#Ahora concatenamos los dummies quitando el stress_level
df_sel = pd.concat([data.drop('Stress_Level',axis=1), df_stress], axis=1)
```
